## Trick to tokenize large datasets faster

We can see that when we use the parameter use_fast = True we get around 7 seconds of execution time and when we keep it to False, we get around 17 seconds of execution time.

This is because when we use use_fast = True the backend code is implemented using Rust language and when we use use_fast = False, python code is run in backend.

```python
import time
text = ["This is a sample sentence for tokenization."] * 100000
```

```python
# Fast method

start = time.time()
tok_fast = AutoTokenizer.from_pretrained("bert-base-cased", use_fast=True)
tok_fast(text, padding=True, truncation=True)
print("Fast time: ", time.time() - start)
```

```python
# Normal / slow method

start = time.time()
tok_slow = AutoTokenizer.from_pretrained("bert-base-cased", use_fast=False)
tok_slow(text, padding=True, truncation=True)
print("Slow time: ", time.time() - start)
```

# Creating Your own Tokenizer from scratch

The most famous algorithm on HuggingFace for creating your own tokenizer is BPE (Byte Pair Encoding) and it is used alongside Hugging Face's fast tokenizer library.

BPE is a subword tokenization algorithm that helps break down words into smaller pieces (subwords) making it useful for handling rarte words and unknown words in NLP tasks.

```python
from tokenizers import Tokenizer, models, pre_tokenizers, trainers
corpus = ["The quick brown fox jumps over the lazy dog.", "Artificial intelligence is transforming the world.", "Hugging Face makes working with NLP easier. \n"]
```

```python
corpus
```

```python
tokenizer = Tokenizer(models.BPE())
```

Removing the white spaces between the words.

```python
tokenizer.pre_tokenizer = pre_tokenizers.Whitespace()
```

```python
trainer = trainers.BpeTrainer(vocab_size=50)
```

```python
tokenizer.train_from_iterator(corpus, trainer)
```

```python
tokenizer.save("custom-tokenizer.json")
```

Now Loading the tokenizer that we created just now.

# Performing small Practical

```python
from torch.nn.functional import cosine_similarity
```

```python
sentence1 = "Hello, how are you?"
sentence2 = "Hi, how are you?"
```

```python
inputs1 = tokenizer(sentence1, return_tensors="pt")
inputs2 = tokenizer(sentence2, return_tensors="pt")
```

```python
with torch.no_grad():
  output1 = model(**inputs1).last_hidden_state[0].mean(dim=0)
  output2 = model(**inputs2).last_hidden_state[0].mean(dim=0)
```

```python
similarity = cosine_similarity(output1,output2,dim=0)
print(f"Similarity: {similarity.item():4f}")
```

# Evaluation Metrics

```python
! pip install evaluate rouge_score evaluate[visualization]
```

```python
import evaluate
```

```python
accuracy = evaluate.load("accuracy")
```

```python
result = accuracy.compute(predictions=[0,1,1,0], references=[0,1,0,0])
```

```python
print(result)
```

```python
def compute_metrics(eval_pred):
  logits, labels = eval_pred
  predictions = logits.argmax(axis=-1)
  return accuracy.compute(predictions=predictions, references=labels)
```

```python
from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer
```

```python
train_dataset = load_dataset("imdb", split="train[:100]")
test_dataset = load_dataset("imdb", split="test[:100]")
```

```python
model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased", num_labels = 2)
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
```

```python
def tokenize(example):
  return tokenizer(example["text"], truncation=True, padding="max_length")
```

```python
train_encoded = train_dataset.map(tokenize, batched=True)
test_encoded = test_dataset.map(tokenize, batched=True)
```

This is the encoded format used for Torch.

```python
train_encoded.set_format("torch", columns=["input_ids", "attention_mask", "label"])
test_encoded.set_format("torch", columns=["input_ids", "attention_mask", "label"])
```

```python
args = TrainingArguments(
    output_dir="eval-check",
    per_device_eval_batch_size=32,
    report_to="none"
)
```

```python
trainer = Trainer(
    model=model,
    args=args,
    compute_metrics=compute_metrics
)
```

```python
metrics = trainer.evaluate(eval_dataset=test_encoded)
```

```python
print(metrics)
```
